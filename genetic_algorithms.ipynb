```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("k_means_output.csv")
```

```python
features = data[["frequency", "latitude", "longitude"]].values
```

```python
features
```

```python
# Extract the cluster indices into a numpy array
cluster_indices = data["Cluster"].values
cluster_indices
     

 
    
```

```python

# Loop over each cluster and extract the corresponding data points

old_cluster = [[], [], [], [], []]
for i in range(cluster_indices.size):
    # Use boolean indexing to extract the data points for this cluster
    old_cluster[cluster_indices[i]].append(i);
clusters = []
for i in range(len(old_cluster)) :
    clusters.append(np.array(old_cluster[i]))
clusters
```

Clusters indexes of initial population

```python
clusters
```

```python
population = features 
population
```

```python
bitPopulation = []
bitCluster = "0"*46

for clusterItem in clusters:
    for loc in clusterItem:
        bitCluster = bitCluster[:loc] + "1" + bitCluster[loc + 1:]
    bitPopulation.append(bitCluster)
    bitCluster = "0"*46
    
bitPopulation
```

Algorithm : 

1. problem definition :-  
a. To find the good clusters among the kmeans cluster i.e which has most frequency in hotspot   
b. A good cluster is represented by which has highest sum of frequency seperation score     

2. population in numpy arrays --> frequency , latitude and longitude , cluster no   
    population - output of kmeans algo

we represented 45 locations as binary string - "1000101000000000000001010"

3.  define fitness function(population , cluster)
     sum of frequencies represent fitness scores
     seperation of longitudes and latitudes

4.  selection 
    - select top 2 fit parents
     
5.  crossover 
     - cross over at random index 
6.  mutation : -
      - less fitness bits flipped 

7.  stop  result plot on the bubble map
